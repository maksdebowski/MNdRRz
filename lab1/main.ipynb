# RÃ³wnanie Prothero- Robinsona

```python
def x_prim(x, t, lam):
    return lam*(x-np.cos(t)) - np.sin(t)

T_0 = 0.0
KSI = 1.0
T = 2.0
N = 50 # na ile czÄÅci dzielimy przedziaÅ
X_0 = 1.0
h = (T- T_0)/N
LAM = -2.0

# DANE DLA COSINUSA
x = np.linspace(0,2,1000)
y = np.cos(x)


# ROZWIÄZANIE
wezly = np.linspace(T_0, T, N+1)
rozwiazania = np.zeros(N+1)
blad = np.zeros(N+1)
lambda_vals = [-2,-10,-50, -100,-1000]
for lam in lambda_vals:
    for i, wezel in enumerate(wezly):
        if i == 0:
            rozwiazania[i] = X_0
            continue
        rozwiazania[i] = rozwiazania[i-1]+ h*x_prim(rozwiazania[i-1], wezel, lam)
        blad[i] = np.abs(rozwiazania[i] - np.cos(wezel))
    plt.figure()
    plt.title(f"Lambda = {lam},  errmax =  {np.max(blad)}")
    plt.scatter(wezly, rozwiazania)
    plt.plot(x,y, color='orange')
    plt.show()
```
