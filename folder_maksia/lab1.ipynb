# Równanie Prothero- Robinsona

```python
def x_prim(x, t, lam):
    return lam*(x-np.cos(t)) - np.sin(t)

T_0 = 0.0
KSI = 1.0
T = 2.0 # na ile części dzielimy przedział
X_0 = 1.0
LAM = -2.0

# DANE DLA COSINUSA
x = np.linspace(0,2,1000)
y = np.cos(x)


# ROZWIĄZANIE
lambda_vals = [-2,-10,-50, -100,-1000]
N = 50
h = (T- T_0)/N

for lambda_ in lambda_vals:
    wezly = np.linspace(T_0, T, N+1)
    rozwiazania = np.zeros(N+1)
    blad = np.zeros(N+1)
    for i, wezel in enumerate(wezly):
        if i == 0:
            rozwiazania[i] = X_0
            continue
        rozwiazania[i] = rozwiazania[i-1]+ h*x_prim(rozwiazania[i-1], wezly[i-1], lambda_)
        blad[i] = np.abs(rozwiazania[i] - np.cos(wezel))
    plt.figure(figsize=(12,4))
    plt.title(f"lambda = {lambda_} errmax = {np.sum(blad)}")
    plt.plot(x,y, color="red")
    plt.scatter(wezly, rozwiazania)
    blad2 = np.abs(np.cos(T) - rozwiazania[-1])
```
