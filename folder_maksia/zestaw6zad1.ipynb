```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def r(z):
    return -(z-1) + np.sqrt((z-1)**2 - 4*z) / 2
```

```python
alpha_vals = np.linspace(-5, 5, 1000) + 1j * np.linspace(-5, 5, 1000)[:, None]
stable = []

for alpha_row in alpha_vals:
    for alpha in alpha_row:
        a = 1
        b = alpha - 1
        c = -alpha

        discriminant = b**2 - 4*a*c
        r1 = (-b + np.sqrt(discriminant)) / (2*a)
        r2 = (-b - np.sqrt(discriminant)) / (2*a)

        if abs(r1) <= 1 and abs(r2) <= 1:
            stable.append(alpha)

stable = np.array(stable)

plt.figure(figsize=(10, 6))
plt.scatter(stable.real, stable.imag, color='red', s=1, label='Stability Region (|r| $\\leq$ 1)')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')

plt.title("Stability Region for Complex $\\alpha$")
plt.xlabel("Re($\\alpha$)")
plt.ylabel("Im($\\alpha$)")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid(True)
plt.show()
```
